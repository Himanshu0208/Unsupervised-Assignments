```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=1000 , noise=0.08)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN
cluster=DBSCAN(eps=0.1,min_samples=5,metric='euclidean')
y_pred=cluster.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
```

- yellow points are the first cluster
- the green points are the second cluster 
- the points in the purple are the outliers/noise
