__True Positive (TP) and False Positive (FP) Rates:__

- True Positive Rate (TPR) or Recall: TPR = TP / (TP + FN), where TP represents the number of correctly detected anomalies and FN is the number of missed anomalies.
- False Positive Rate (FPR): FPR = FP / (FP + TN), where FP represents the number of false alarms (normal instances incorrectly classified as anomalies) and TN is the number of correctly classified normal instances.

- Precision: Precision = TP / (TP + FP), representing the proportion of correctly detected anomalies among all detected instances.
- Accuracy: Accuracy = (TP + TN) / (TP + FP + TN + FN), indicating the overall correctness of the anomaly detection results.

__Local Outlier Detection:__

- Intrusion Detection: In network security, local outlier detection is often more suitable. It focuses on identifying anomalies within specific network connections or nodes, such as detecting unusual behavior or network traffic patterns that deviate from the norm.

- Fraud Detection: Local outlier detection is effective for detecting anomalies at a transaction level. It can identify unusual activities within a specific account or a subset of transactions, such as detecting fraudulent credit card transactions or money laundering.

- Sensor Networks: Local outlier detection is commonly used in sensor networks to identify anomalous readings or behavior of individual sensors. It helps detect faulty or malfunctioning sensors that produce abnormal data compared to their neighboring sensors.

__Global Outlier Detection:__

- Manufacturing Quality Control: Global outlier detection is often employed in manufacturing processes to identify defective products or faulty equipment that affect the entire production line. It focuses on detecting anomalies that occur across the entire system rather than on individual instances.

- Finance and Stock Market: Global outlier detection is useful for detecting anomalies that affect the entire market or financial system. It can identify major events like stock market crashes, economic downturns, or financial fraud that impact multiple stocks, indices, or financial indicators.

- Health Monitoring: Global outlier detection can be used in health monitoring to identify widespread anomalies or disease outbreaks. It helps detect patterns that deviate from normal health conditions across a population, such as identifying disease clusters or detecting abnormal health trends.
